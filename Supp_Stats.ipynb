## Confidence Intervals

```python
x.head()
```

### Confidence Interval - Mean

```python
import scipy.stats as stats
```

```python
stats.t.interval(0.95, df=len(x["Loss"])-1,
              loc=np.mean(x["Loss"]),
              scale=stats.sem(x["Loss"]))
```

Q Q Plot

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
fig = sm.qqplot(x["Loss"], stats.t, fit=True, line='45')
plt.show()
```

## Hypothesis Tests

### One Sample T Test

```python
x4 = pd.read_excel('/content/drive/MyDrive/BI BA Handbook/Appendix/statistics.xlsx', sheet_name="GRE")
```

```python
x4.head()
```

```python
from scipy.stats import ttest_1samp
```

```python
ttest_1samp(x4["GRE Score"], popmean=400, alternative="less")
```

### One Sample Proportions Test

```python
import statsmodels.stats.proportion as prop
```

```python
prop.proportions_ztest(550, 1000, 0.5, alternative="smaller")
```

### Pooled T-Test

```python
x5 = pd.read_excel('/content/drive/MyDrive/BI BA Handbook/Appendix/statistics.xlsx', sheet_name="GPA")
```

```python
x5.head()
```

```python
from scipy.stats import bartlett
```

```python
bartlett(x5["Own Car"], x5["Dont Own Car"])
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(x5["Own Car"], x5["Dont Own Car"], equal_var=True, alternative="less")
```
