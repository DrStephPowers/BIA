```python
def ICR(revenue, var, fix, interest_pmt): #interest coverage ratio
    EBIT = revenue * (1-var)-fix
    ICR_calc = EBIT / interest_pmt
    return ICR_calc
```

## **4. Fraud Detection**

***Predictive Analytics*** - What could happen to the organization?

### Benford's Law

Predict whether transactions are fraudulent

```python
transactions.head()
```

```python
dig = transactions["amount"].astype(str).str[:1]
dig = dig[dig != '-']
dig = dig.apply(int)
```

```python
freq = dig.value_counts().sort_index()
freq
```

```python
plt.plot(freq.index, freq.values)
plt.xticks([])
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.interpolate import make_interp_spline
```

```python
def smoothify(yInput, topwordlist):
    x = np.array(range(0,len(topwordlist.index)))
    y = np.array(yInput)
    x_smooth = np.linspace(x.min(), x.max(), 600)
    spl = make_interp_spline(x, y, k=3)
    y_smooth = spl(x_smooth)
    return x_smooth, y_smooth
```

```python
ziffianCurveValues = [100/i for i in range(1, len(freq.index)+1)]
```

```python
x, y = smoothify(ziffianCurveValues, freq)
```

```python
freq_per = freq.values/freq.values[0]*100
```

```python
plt.plot(freq.index, freq_per)
plt.plot(x+1,y)
plt.legend(["Numbers", "Benford's Law"])
plt.ylabel("Percent Frequency")
plt.xlabel("Index")
plt.show()
```

### Prophet Model (forecasting with seasonality)

Identify misuse of employee loyalty program

```python
discounts.head()
```

```python
!pip install prophet
```

```python
from prophet import Prophet
from seaborn import scatterplot
import matplotlib.pyplot as plt
import numpy as np
```

```python
scatterplot(x=discounts["date"],y=discounts["discount_amount"])
plt.xlabel('Date')
plt.ylabel('Discount Amount')
plt.xticks([])
plt.show()
```

```python
run = discounts[["date", "discount_amount"]].copy()
```

```python
run.rename(columns={"date": "ds", "discount_amount": "y"}, inplace=True)
```

```python
model = Prophet(interval_width=0.95, yearly_seasonality=True, weekly_seasonality=True)
```

```python
model.fit(run)
```

```python
forecast = model.predict(run)
forecast.head()
```

```python
performance = pd.concat([discounts, forecast], axis=1)
```

```python
performance['anomaly'] = performance.apply(lambda rows: 1 if ((rows.discount_amount<rows.yhat_lower)|(rows.discount_amount>rows.yhat_upper)) else 0, axis = 1)
```

```python
performance['color'] = np.where(performance['anomaly']== 1, 'red', 'green')
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.update_layout(
    height=800,
    title="Anomaly Detection",
    xaxis_title="Date",
    yaxis_title="Discount Amount",

)

fig.add_trace(go.Scatter(x=performance['ds'], y=performance['discount_amount'],
                    mode='markers',
                    name='Anomalies - Red Colored Dots',
                    marker_color=performance.color,

                        ))
fig.add_trace(go.Scatter(x=performance['ds'], y=performance['yhat'],
                    mode='lines',
                    name='Views',
                    marker_color="black"
                        ))
fig.show()
```

## ***5. Late Payments***

Classification Model (Decision Tree)

```python
distress.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
x_train, x_test, y_train, y_test = train_test_split(distress[["current_ratio", "debt_to_equity", "return_on_assets"]],
distress[["actual_status"]], test_size=0.2, stratify=distress[["actual_status"]])
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(x_train, y_train)
```

```python
y_pred = dtree.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtree.classes_)
disp.plot()
```

```python
tree.plot_tree(dtree)
plt.show()
```
