## Mode

Mode from Original Dataset

Axis = 0 indicates that the mode is found for each column. Axis = 1 would give the mode by row.

Mode can be restricted to only numeric columns. If you have coded the survey, then use numeric_only=TRUE so that it doesn't show the open ended responses. The reason the information below contains rows of NaN is because it is showing each unique response for the open ended questions (all responses tie for a mode of 1).

```python
df.mode(axis=0, numeric_only=False, dropna=True)
```

Mode from Coded Data

Notice that the mode for Q2 previously was Full Time. Recall that Part Time was coded as 1 and Full Time was coded as 2. Below the mode for Q2 is 2.

An additional line of code was added to show only the first row. The 62 rows of NaN aren't needed.

A third line of code allows for the data to be transposed. Otherwise, the varibles are listed as columns, which is inconsistent with the other descriptive statistics.

```python
df2_mode1 = df2.mode(axis=0, numeric_only=True, dropna=True)
df2_mode2 = df2_mode1.head(1)
df2_mode = df2_mode2.T
print(df2_mode)
```

## Summary Statistics

Creating a table of Summary Statitics

```python
df2_mode1 = df2.mode(axis=0, numeric_only=True, dropna=True)
df2_mode2 = df2_mode1.head(1)
df2_mode = df2_mode2.T

df2_range = df2.max()-df2.min()

Q1 = df2.quantile(0.25)
Q3 = df2.quantile(0.75)
df2_IQR = Q3-Q1

df_summary = pd.DataFrame()
df_summary["Min"] = df2.min()
df_summary["Max"] = df2.max()
df_summary["Mode"] = df2_mode
df_summary["Median"] = df2.median()
df_summary["Mean"] = df2.mean()
df_summary["Range"] = df2_range
df_summary["IQR"] = df2_IQR
df_summary["Std Dev"] = df2.std()
print(df_summary)
```
