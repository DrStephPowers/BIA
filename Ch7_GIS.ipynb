```python
data = gpd.read_file('/content/drive/MyDrive/BI BA Handbook/Chapter 7/maps/3rdfloor2025.shx')
```

```python
data.head()
```

```python
import plotly.express as px

data["data_ua17"] = data["data_ua17"].astype(float)
```

```python
fig = px.choropleth(data, geojson=data.geometry,
                    locations=data.index, color="data_ua17",
                    height=500,
                   color_continuous_scale="Viridis")
fig.update_geos(fitbounds="locations", visible=True)
fig.update_layout(
    title_text='Utilization of 3rd Floor'
)
fig.update(layout = dict(title=dict(x=0.5)))
fig.show()
```
