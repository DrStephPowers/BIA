```python
%%writefile app.py
import streamlit as st
import pandas as pd
import altair as alt
import plotly.express as px
from datetime import date

st.set_page_config(
    page_title="Youtube Analytics",
    page_icon="📈",
    layout="wide",
    initial_sidebar_state="expanded")

alt.themes.enable("dark")

################################
# Load data

TotalViews = pd.read_csv('/content/drive/MyDrive/BI BA Handbook/Chapter 7/youtube/Totalviews.csv')
age = pd.read_csv('/content/drive/MyDrive/BI BA Handbook/Chapter 7/youtube/age.csv')
geogfiltered = pd.read_csv('/content/drive/MyDrive/BI BA Handbook/Chapter 7/youtube/geogfiltered.csv')
gender = pd.read_csv('/content/drive/MyDrive/BI BA Handbook/Chapter 7/youtube/gender.csv')
content = pd.read_csv('/content/drive/MyDrive/BI BA Handbook/Chapter 7/youtube/content.csv')
content = content.tail(-1)

################################
# Sidebar
with st.sidebar:
    st.title('Analytics Dashboard')

    yearslide = st.slider("Yearly Data", 2013, 2024, 2020)
    st.write("Year:", yearslide)

    yearrange = st.slider("Select a range of years", 2013, 2024, (2018, 2022) )
    st.write("Year Range:", yearrange)

    agegroup = st.selectbox('Select an age group', age['Viewer age'])
    st.write('Age Group:', agegroup)

    gendergroup = st.selectbox('Select a gender', gender['Viewer gender'])
    st.write('Gender Group:', gendergroup)

    color_theme_list = ['blues', 'cividis', 'greens', 'inferno', 'magma',
                        'plasma', 'reds', 'rainbow', 'turbo', 'viridis']
    selected_color_theme = st.selectbox('Select a color theme',
                                        color_theme_list)

    single_color_list = ['blue', 'green', 'orange', 'red']
    selected_color = st.selectbox('Select a color', single_color_list)


#######################
# Plots
def make_choropleth(input_df, input_id, input_column, input_color_theme):
    choropleth = px.choropleth(input_df, locations=input_id, color=input_column,
                               color_continuous_scale=input_color_theme,
                               hover_name= input_id,
                               locationmode='ISO-3')
    choropleth.update_layout(
        template='plotly_dark',
        plot_bgcolor='rgba(0, 0, 0, 0)',
        paper_bgcolor='rgba(0, 0, 0, 0)',
        margin=dict(l=0, r=0, t=0, b=0),
        height=350
    )
    return choropleth

# Line chart
def make_linechart(input_df, input_x, input_y):
  YearX = pd.to_datetime(input_df[input_x])
  input_df["Years"] = YearX.dt.year

  df_range = input_df[input_df["Years"].between(yearrange[0], yearrange[1], inclusive='both')]
  linechart =  st.line_chart(df_range, x=input_x, y=input_y, x_label="Date",
                             y_label="Viewership")
  return linechart

# Donut chart
def make_donut(input_response, input_text, input_color):
  if input_color == 'blue':
      chart_color = ['#29b5e8', '#155F7A']
  if input_color == 'green':
      chart_color = ['#27AE60', '#12783D']
  if input_color == 'orange':
      chart_color = ['#F39C12', '#875A12']
  if input_color == 'red':
      chart_color = ['#E74C3C', '#781F16']

  source = pd.DataFrame({
      "Topic": ['', input_text],
      "% value": [100-input_response, input_response]
  })

  plot = alt.Chart(source).mark_arc(innerRadius=50).encode(
      theta="% value",
      color= alt.Color("Topic:N",
                      scale=alt.Scale(
                          #domain=['A', 'B'],
                          domain=[input_text, ''],
                          # range=['#29b5e8', '#155F7A']),  # 31333F
                          range=chart_color),
                          legend=None),
  ).properties(width=130, height=130)
  text = plot.mark_text(align='center', color="#29b5e8", font="Lato", fontSize=32, fontWeight=700, fontStyle="italic").encode(text=alt.value(f'{input_response} %'))
  return plot + text


#######################
# Dashboard Main Panel
col = st.columns((1.5, 4.5, 2), gap='medium')

with col[0]:
    st.markdown('#### Number of Viewers')

    Year1 = pd.to_datetime(TotalViews["Date"])
    TotalViews["Year"] = Year1.dt.year

    df_select_year = TotalViews[TotalViews["Year"] == yearslide]

    if yearslide > 2013:
        viewtotal = df_select_year["Views"].sum()
        yearbefore = yearslide - 1
        df_select_year_before = TotalViews[TotalViews["Year"]
          == yearbefore]
        viewtotal_before = df_select_year_before["Views"].sum()
        viewtotal_delta = viewtotal - viewtotal_before
    else:
        viewtotal = df_select_year["Views"].sum()
        viewtotal_delta = ''
    st.metric(label=str(yearslide), value=int(viewtotal), delta=int(viewtotal_delta))

    st.markdown('#### Viewer Demographics')

    donut1 = gender.loc[gender['Viewer gender'] == gendergroup, 'Watch time (hours) (%)'].iloc[0]
    donut_gender = make_donut(donut1, gendergroup, selected_color)

    donut2 = age.loc[age['Viewer age'] == agegroup, 'Watch time (hours) (%)'].iloc[0]
    donut_age = make_donut(donut2, agegroup, selected_color)

    donuts_col = st.columns((0.2, 1, 0.2))
    with donuts_col[1]:
        st.write(gendergroup)
        st.altair_chart(donut_gender)
        st.write(agegroup)
        st.altair_chart(donut_age)

with col[1]:
    st.markdown('#### Viewership')

    choropleth = make_choropleth(geogfiltered, 'iso_alpha', 'Views',
                                 selected_color_theme)
    st.plotly_chart(choropleth, use_container_width=True)

    linechart = make_linechart(TotalViews, 'Date', 'Views')

with col[2]:
    st.markdown('#### Top Videos')

    st.dataframe(content[["Video title", "Views"]],
                 column_order=("Video title", "Views"),
                 hide_index=True,
                 width=None,
                 column_config={
                    "Video title": st.column_config.TextColumn(
                        "Video title",
                    ),
                    "Views": st.column_config.ProgressColumn(
                        "Views",
                        format="%f",
                        min_value=0,
                        max_value=content["Views"].max(),
                     )}
                 )

    with st.expander('About', expanded=True):
        st.write('''
            - Data: Powers Youtube Channel
            ''')
```
